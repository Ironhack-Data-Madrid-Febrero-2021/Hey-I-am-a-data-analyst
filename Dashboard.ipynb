# We would like to know which films are the best to start offering 

```python
# Creating a list in order to compare them
netflix_films = netflix['Title'].to_list()
imdb_ = imdb['English_Title'].to_list()
```

```python
# We create a function for cleaning English titles.
def strip_(list):
    imdb_films = []
    for film in list:
        x = film.strip()
        imdb_films.append(x)
    return imdb_films

# Save the clean titles in a list
imdb_films = strip_(imdb_)
```

```python
no_netflix = []
#if 101 not in l: l.append(101)'
def free_films(films):
    for film in films:
        if film not in netflix_films: no_netflix.append(film)
    return no_netflix

target_films = free_films(imdb_films)
```

```python
print("Some of the films The company can start offering are: ", target_films[:10])
```
