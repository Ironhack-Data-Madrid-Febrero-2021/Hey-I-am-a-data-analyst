### We create a new database to prove the performance of our model

```python
new_db=db[['Name', 'Position', 'Score']]
```

```python
new_db
```

### Model results

```python
#Here we have the results of our model, as we can see there are cheap and expensive players. 
#The most attractive fact about our model is that you don't have to spend a great amount of money to sign a good player
#You only have to get the variables election right, those that your team really need
```

```python
#With our code, and the variables you ask for, we suggest to sign one of these goalkeepers
```

```python
Goalkeeper = new_db.loc[(new_db['Position'] == 'Goalkeeper') & (new_db['Score'] > 70) ]
Goalkeeper
```

```python
Goalkeeper_graph=Goalkeeper.groupby(['Name','Score'])
Goalkeeper_graph

Goalkeeper_graphic= Goalkeeper.plot.bar(x='Name', y= 'Score', figsize=(12,4), color='Red')
```

```python
#With our code, and the variables you ask for, we suggest to sign one of these defenders
```

```python
Defender = new_db.loc[(new_db['Position'] == 'Defender') & (new_db['Score'] > 70) ]
Defender

```

```python
Defender_graph=Defender.groupby(['Name','Score'])
Defender_graph

Defender_graphic= Defender.plot.bar(x='Name', y='Score', figsize=(12,4), color='Blue')
```

```python
#With our code, and the variables you ask for, we suggest to sign one of these midfielders
```

```python
Midfielder= new_db.loc[(new_db['Position'] == 'Midfielder') & (new_db['Score'] > 180) ]
Midfielder
```

```python
Midfielder_graph=Midfielder.groupby(['Name','Score'])
Midfielder_graph

Midfielder_graphic= Midfielder.plot.bar(x='Name', y='Score', figsize=(12,4), color='Green')
```
